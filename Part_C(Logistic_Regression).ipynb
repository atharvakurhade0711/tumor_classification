```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('data_set.csv')
raw_data
```

```python
dataset1 = raw_data.drop('id', axis=1)
dataset1 = dataset1.dropna()
dataset1.diagnosis.replace(['M', 'B'], [1, 0], inplace=True)
dataset1
```

```python
dataset2 = raw_data.drop('id', axis=1)
dataset2.diagnosis.replace(['M', 'B'], [1, 0], inplace=True)

for column in dataset2.columns[1:]:
    dataset2[column] = (dataset2[column] - dataset2[column].mean())/dataset2[column].std()
    
dataset2.fillna(0, inplace=True)
dataset2
```

```python
def sigmoid(x):
    if x>=0:
        return 1/(1+np.exp(-x))
    else:
        return np.exp(x)/(1+np.exp(x))
    
def err(training_set, w):
    e=0
    for example in training_set:
        x = example[1:]
        y = example[0]
        h = sigmoid(w.dot(x))
        if (h==1):
            h = 1 - 10**(-16)
        elif (h==0):
            h = 10**(-16)
            
        e -= y*np.log(h) + (1-y)*(np.log(1-h)) 
    
    return e
```

```python
def grad(training_set, w):
    g = 0
    for example in training_set:
        x = example[1:]
        y = example[0]
        h = sigmoid(w.dot(x))
        g += (y-h)*x
    
    return g

def batch_grad(training_set):
    w = np.zeros(30)
    e1 = err(training_set, w)
    e2 = 0
    cost_list = []
    while e1==e2 or np.abs(e1-e2)>0.001:
        e1 = err(training_set, w)
        w += 0.0001*grad(training_set, w)
        e2 = err(training_set, w)
        cost_list.append(e2)
    
    return w, cost_list
```

```python
def stoch_grad(training_set):
    w = np.zeros(30)
    e1 = err(training_set, w)
    e2 = 0
    cost_list = []
    while e1<=e2 or np.abs(e1-e2)>0.000000001:
        e1 = err(training_set, w)
        example = training_set[np.random.randint(len(training_set))]
        x = example[1:]
        y = example[0]
        h = sigmoid(w.dot(x))
        w += 0.01*(y - h)*x
        e2 = err(training_set, w)
        cost_list.append(e2) 
    
    return w, cost_list
```

```python
def func(x):
    if x>=0.5:
        return 1
    else:
        return 0
    
def evaluate(testing_set, w):
    count =0
    for example in testing_set:
        x = example[1:]
        y = example[0]
        h = sigmoid(w.dot(x))
        if y == func(h):
            count +=1
        
    return count
```

```python
training_data = dataset1.sample(frac=2/3)
testing_data = dataset1.drop(training_data.index)
training_data
```

```python
testing_data
```

```python
training_set = training_data.to_numpy()
testing_set = testing_data.to_numpy()
w, cost = batch_grad(training_set)
cost
```

```python
plt.figure(figsize = (9, 6))
plt.plot(cost)
plt.xlabel('Iterations', fontsize = 14)
plt.ylabel('Cost', fontsize = 14)
plt.tight_layout()
plt.show()
```
